```python
import numpy as np
from gensim.models import Word2Vec

# Example synthetic user-item interaction data
user_item_data = [
    ('127', ['4447', '8622', '7147']),
    ('190', ['8622', '7147']),
    ('383', ['4447', '7147']),
    # Add more user-item interactions as needed
]

# Train Word2Vec model to learn user embeddings
user_embeddings_dim = 100
window_size = 5
min_count = 1

# Prepare data for Word2Vec training
sentences = [user_items[1] for user_items in user_item_data]

# Train Word2Vec model
w2v_model = Word2Vec(sentences, vector_size=user_embeddings_dim, window=window_size, min_count=min_count)

# Prepare embedding file
embedding_file_path = 'user_embeddings.useremb'

with open(embedding_file_path, 'w') as file:
    # Write header
    file.write('uid:token\n')
    file.write('user_emb:float_seq\n')
    
    # Write user embeddings
    for user_id in w2v_model.wv.index_to_key:
        embedding_vector = ' '.join(map(str, w2v_model.wv[user_id]))
        file.write(f'{user_id}\n')
        file.write(f'{embedding_vector}\n')

print(f'User embeddings saved to: {embedding_file_path}')
```

```python
import os
from recbole.config import Config
from recbole.data import create_dataset, data_preparation
from recbole.quick_start import run_recbole
from recbole.model.general_recommender import BPR
from recbole.trainer import Trainer
from recbole.utils import init_seed, init_logger
#from recbole.model.general_recommender import GeneralRecommender

# Step 1: Organize the Data
data_folder = '/Users/bandhaviparvathaneni/Downloads/Recbole'

# Step 2: Create a Configuration File
config_file_path = '/Users/bandhaviparvathaneni/Downloads/Recbole/recbole.yaml'

for file_name in os.listdir(data_folder):
    if os.path.isfile(os.path.join(data_folder, file_name)):
        print(file_name)
        
# Step 3: Load Configuration 
print("Loading configuration from file:", config_file_path)
config = Config(model='BPR', dataset='ml-100k', config_file_list=[config_file_path])


# Step 4: Create Dataset
print("Creating dataset...")
dataset = create_dataset(config)
print("Dataset created successfully.")

# Step 5: Prepare Data
print("Preparing data...")
train_data, valid_data, test_data = data_preparation(config, dataset)
print("Data preparation complete.")

# Step 6: Run Models
print("Running models...")
run_recbole(dataset='ml-100k', model=BPR)

print("Models run successfully.")
```
